```python
import pandas as pd
data = pd.read_csv(r"C:\Users\DELL\OneDrive\Documents\python\python ML diabetes\diabetes.csv")
data.head()
```

```python
data.info()
```

```python
x=data.drop(["Outcome"],axis=1)
x
```

```python
y=data["Outcome"]
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train
```

```python
y_train
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
print(y_pred)
#data.head(50)
```

```python
import joblib
joblib.dump(model,"diab_pred.pkl")
```

```python
out={0:"No",1:"Yes"}
load_model=joblib.load('diab_pred.pkl')
y_pred=load_model.predict([[6,148,72,35,0,33.6,0.627,50]])
out[y_pred[0]]
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
accuracy = accuracy_score(y_test,y_pred) 
accuracy
```
